```python
benzine = ["Gasoline", "Super 95","Regular/Benzine 91", "Regular/Benzine E10 91","Super E10 95","Super Plus 98","Super Plus E10 98", "Ethanol", "Others"]
lpg = ["LPG","Liquid petroleum gas", "CNG", "Biogas", "Domestic gas H", "Domestic gas L", "Hydrogen"]
dual = ["Domestic gas H / Super E10 95 / Super Plus E10 98 / Super 95 / Super Plus 98 / Domestic gas L", 
              "Domestic gas L / Super 95 / Domestic gas H",
              "Super 95 / Super Plus 98 / Liquid petroleum gas",
              "Super 95 / Liquid petroleum gas"]
def fueltype(x):
    if x in benzine:
        return "Benzine"
    elif x in lpg:
        return "LPG/CNG"
    elif x in dual:
        return "Benzin/LPG"
    else:
        return x
df0["Fuel_Type"] = df0.Fuel_Type.apply(fueltype)
```

```python
df0['Fuel_Type'] = df0['Fuel_Type'].replace("None",np.nan)
df0['Fuel_Type'] = df0['Fuel_Type'].replace("nan",np.nan)
df0['Fuel_Type'] = df0['Fuel_Type'].replace("NaN",np.nan)
df0['Fuel_Type'] = df0['Fuel_Type'].replace("",np.nan)
```

```python
df0["Fuel_Type"].value_counts(dropna = False)
```

```python
for group in list(df0["Make_Model"].unique()):
    cond = df0["Make_Model"] == group
    mode = list(df0[cond]["Fuel_Type"].mode())
    if mode != []:
        df0.loc[cond, "Fuel_Type"] = df0.loc[cond, "Fuel_Type"].fillna(df0[cond]["Fuel_Type"].mode()[0])
    else:
        df0.loc[cond, "Fuel_Type"] = df0.loc[cond, "Fuel_Type"].fillna(df0["Fuel_Type"].mode()[0])
    
```

```python
def Engine_Size_Converter(x):
    if float(x) is np.nan :
        return np.nan
    elif int(x) == 1 :
        return 1000
    elif int(x) == 2 or float(x) == 20 :
        return 2000
    elif int(x) == 3 :
        return 3000
    elif int(x) == 4 :
        return 4000
    elif int(x) == 5 :
        return 5000
    elif int(x) <= 1100 :
        return 1000
    elif int(x) <= 1300 :
        return 1200
    elif int(x) <= 1500 :
        return 1400
    elif int(x) <= 1800 :
        return 1600
    elif int(x) <= 2500 :
        return 2000
    elif int(x) <= 3500 :
        return 3000
    elif int(x) <= 4500 :
        return 4000
    elif int(x) <= 5500 :
        return 5000
    elif int(x) > 5500 :
        return 6000
```

```python
df0.Engine_Size = df0.Engine_Size.apply(Engine_Size_Converter)
```

```python
df0.Engine_Size.value_counts(dropna=False)
```
